```python
def Grading(row):
    if row["Total"]> 300:
        row["Grade"]="A+"
    elif row["Total"]>250:
        row["Grade"]="A"
    elif row["Total"]>200:
        row["Grade"]="B"
    return row
        
        
```

```python
dd=dd.apply(Grading, axis=1)
```

```python
dd
```

```python
def FP(row):
    if (row.loc["A":"E"]>35).all():
        row["Res"]="Pass"
    else:
        row["Res"]="Fail"
    return row
```

```python
dd= dd.apply(FP, axis=1)
```

```python
df= pd.read_csv("social_media_engagement.csv")
```

```python
df
```

```python
def review(row):
    if row["likes"]>=2000:
        row["Reach"]="Good"
    else:
        row["Reach"]="Bad"
    return row 
    
```

```python
df= df.apply(review, axis=1)
```

```python
df
```

```python
def review1(row):
    if row["likes"]>=500 and row["comments"]<=100:
        return "High"
    elif row["likes"]>2000:
        return "Medium"
    elif row["shares"]>1000:
        return "Low"
```

```python
df["Risk"]= df.apply(review1, axis=1)
```

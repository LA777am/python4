```python
meanhp=dd["horsepower"].mean()
```

```python
dd["horsepower"] = dd["horsepower"].replace("?", np.nan)
```

```python
dd["horsepower"] = pd.to_numeric(dd["horsepower"])
```

```python
dd["horsepower"] = dd["horsepower"].fillna(meanhp)
```

```python
pd.plotting.parallel_coordinates(dd,"cylinders",["horsepower","weight","mpg"])
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df= pd.DataFrame({"dep":["he","it","hr","it","hr"],
                 "age": [20,25,30,22,10],
                 "sal":[100,500,200,300,100],
                 "sex":["F","M","M","F","F"]})
```

```python
df
```

```python
df["age"].hist()
```

```python
px.histogram(data_frame=df, x="age", color="dep", nbins=10, pattern_shape="dep")
```

```python
px.histogram(data_frame=df, x="dep",y="sal",histfunc="count", color="sex")
```

```python
import networkx as nx
```

```python
g= nx.DiGraph()
g.add_node(1)
g.add_node(2)
g.add_node(3)
g.add_node(4)
g.add_edge(1,2)
g.add_edge(2,3)
g.add_edge(3,4)
g.add_edge(4,1)
nx.draw(g, with_labels=True, node_size=1000, edge_color="blue", node_color="red")
```

```python
g= nx.DiGraph()
g.add_nodes_from([1,2,3,4])
g.add_edges_from([(1,2),(2,3),(3,4),(4,1)])
nx.draw(g, with_labels=True, node_size=1000)
```

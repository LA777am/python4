```python
from bs4 import BeautifulSoup
import pandas as pd 
```

```python
with open("t3_vha/Certified used Mercedes-Benz for sale.html", "r") as f:
    data= f.read()
soup= BeautifulSoup(data, "html.parser")
    
```

```python
common_div = soup.find_all("div", class_="vehicle-details")
len(common_div)
```

```python
year=[]
name=[]
prices=[]
mileage=[]
review=[]
rating=[]
for i in common_div:
    tittle= i.find("h2", class_="title").text
    tittle1= tittle.split(" ")
    yer= int(tittle1[0])
    nme=" ".join(tittle1[1:])
    nme1= nme.replace("[","")
    nme2= nme.replace("]","")
    mil = i.find("div", class_="mileage").text
    mil2= int(mil.split(" ")[0].replace(",", ""))
    price = i.find("span", class_="primary-price").text
    price2= price.replace(",","")
    price3= price2.replace("$", "")
    price4=int(price3)
    rat1 = i.find("span", class_="sds-rating__count")
    if rat1:
        rat2= rat1.text
        rat3= float(rat2)
    rev= i.find("span", class_= "sds-rating__link sds-button-link")
    if rev:
        rev1= rev.text
        rev2=rev1.split(" ")
        rev3= rev2[0]
        rev4= rev3.replace("(", "")
        rev5= rev4.replace(",", "")
        rev6= int(rev5)
    review.append(rev6)
    rating.append(rat3)
    name.append(nme2)
    year.append(yer)
    mileage.append(mil2)
    prices.append(price4)
df= pd.DataFrame({"name":name, "Prices":prices, "Mileage":mileage, "year":year, "Rating":rating, "Reviews":review})
df
```

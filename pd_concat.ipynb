```python
df1= pd.DataFrame({"name":["vha", "psp","mss","vby","dps"],
                  "sub":["python","fsd","coa","dm","toc"], 
                  "id":[3,2,1,4,6]})
```

```python
df2=pd.DataFrame({"id":[1,2,3,4,5], "mark":[9,8,10,8,8]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on="id")
```

```python
pd.merge(df1,df2,on="id", how="outer")
```

```python
pd.merge(df1,df2,on="id",how="left")
```

```python
pd.merge(df1,df2,on="id",how="cross")
```

```python
pd.merge(df1,df2,left_index=True, right_index=True, suffixes=("_df1", "_df2"))
```

```python
df1= pd.DataFrame({"name":["vha", "psp","mss","vby","dps"],
                  "sub":["python","fsd","coa","dm","toc"], 
                  "id":[3,2,1,4,6]})
df2=pd.DataFrame({"id":[1,2,3,4,5], "mark":[9,8,10,8,8],"fac":["vha","nas", "vby", "mss","dps"]})
```

```python
pd.merge(df1,df2,left_on="name", right_on="fac", how="outer",suffixes=("_b1_sem3" ,"_b1_sem4"))
```
